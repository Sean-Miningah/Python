```python
import pandas as pd 

mydataset = {
    'cars': ["BMW","Volvo","Ford"],
    'passings': [3,7,2]
}

myvar = pd.DataFrame(mydataset)

print(myvar)
```

A _panda_ Series is like a column in a table.
It is a one_dimensional array holding data of any type.

```python
a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
```

#### Creating Labels
With the __index__ argument, you can name your own labels

```python
a = [1, 7, 2]
myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)
```

When you create labels you can access an tiem by referring to the label.

```python
print(myvar['x'])
```

#### Key/Value Objects as Series
You can also use a key/value object, like a dictionary, when creating a Series

```python
calories = {
    "day1" : "420",
    "day2" : "380", 
    "day3" : 390
}

myvar = pd.Series(calories)
print(myvar)

# The key values become the labels of the dataset
```

```python
# Create a Series using only data from "day1" and "day2"
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index=["day1", "day2"])

print(myvar)
```

## DataFrames

Data sets in Pandas are usually multi-dimensional tables, called DataFrames.
Series is like a column, a DataFrame is the whole table.
