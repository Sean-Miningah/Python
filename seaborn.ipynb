```python
# The barplot below shows the survivors of the titanic crash based on category

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context('paper')

# load dataset
titanic = sns.load_dataset('titanic')
# create plot
sns.barplot(x='embark_town', y='age', data=titanic,
            palette='PuRd', ci=None
            )
plt.legend()
plt.show()
print(titanic.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# load dataset
titanic = sns.load_dataset('titanic')
# create plot
sns.barplot(x='sex', y='survived', hue='class', data=titanic,
            palette='PuRd',
            order=['male', 'female'],
            capsize=0.05,
            saturation=8,
            errcolor='gray', errwidth=2,
            ci='sd'
            )
plt.legend()
plt.show()
```

```python
# To draw a horizontal plot pass 'h' to the parameter, orient of the barplot function as shown below:
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context('paper')
sns.barplot(x='age', y='embark_town', data=titanic,
            palette='PuRd', orient='h',
            )
plt.show()
```

```python
# Count Plot
# The count plot can be thought of as a histogram across a categorical variable. The example below demonstrates the count plot
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context('paper')

# load dataset
titanic = sns.load_dataset('titanic')
# create plot
sns.countplot(x='class', hue='who', data=titanic, palette='magma')
plt.title('Survivors')
plt.show()
```

## KDE PLOT

```python
# KDE plot is a Kernel Density Estimate that is used for visualizing the Probabiliy Density of the 
# continuous or non-paremetric data variables ie we can plot univariate or multiple varialbles together.
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("dark")
iris = sns.load_dataset("iris")
# Plotting the KDE Plot
sns.kdeplot(iris.loc[(iris['species'] == 'setosa'),
            'sepal_length'], color='b', shade=True, Label='setosa')
sns.kdeplot(iris.loc[(iris['species'] == 'virginica'),
            'sepal_length'], color='r', shade=True, Label='virginica')
```

* How to add a legend to a graph?

```python
# Setting up the samples
iris_setosa = iris.query("species=='setosa'")
iris_virginica = iris.query("species=='virginica'")

# Plotting the KDE Plot
sns.kdeplot(iris_setosa['sepal_length'],
            iris_setosa['sepal_width'],
            color='r', shade=True, Label='Iris_Setosa',
            cmap="Reds", shade_lowest=False)
# Plotting the KDE Plot
sns.kdeplot(iris_virginica['sepal_length'],
            iris_virginica['sepal_width'],
            color='r', shade=True, Label='iris_virginica',
            cmap="Blues", shade_lowest=False)
```

## Box plot

```python
# Used for depicting groups of numerical data through the quartiles. It is known as the box whisker diagram because it is composed
# A box plot is composed of a summary of 5 different data points: the minimum, first quartile, median, third quartile and maximum 

import seaborn as sns
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips)
```

> The bottom black horizontal line of the box plot is the minimum value


> The first black horizontal line of the rectangle shape of the box plot is the first quartile
    or 25%


> The second black horizontal line of the rectangle shape of the box plot is Second quartile or 50% or median.


> The third black horizontal line of rectangle shape of the same box plot is third quartile or 75%


> The top black horizontal line of the rectangle shape of the box plot is the maximum value.


> The small diamond shape of the box plot is outlier data.

## Violin Plot

```python
# Violin plots also like boxplots summarise numeric data over a set of categories.
# They are essentially a box plot with a kernel density estimate(KDE) overlaid along with the ranges
# of the box and reflected to make it look nice. 

import seaborn as sns
tips = sns.load_dataset("tips")
ax = sns.violinplot(x=tips["total_bill"])
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker",
               data=tips, palette="muted")
```

```python
## We can also represent the above plot like this by setting the parameter split as True
sns.violinplot(x="day", y="total_bill", hue="smoker",
               data=tips, palette="muted", split=True)
```

## Heatmap

```python
# Is a two-dimensional graphical representation of data where individual values that are contained in a matrix
# are represented as colours

# Now if we get the data of the dataset 'flights' and transform it by monthly as shown below,it 
# can give us a lot of information about the data. But this information is in tabular form and can be better
# displayed by using  heatmap as shown below.

flights = sns.load_dataset("flights")
print(flights)

flights = flights.pivot("month", "year", "passengers")
print(flights)
```

```python
sns.heatmap(flights,linewidths=.5,cmap="YlGnBu")
```

```python
# We can put the respective values in the boxes using the annot paremeter of this function

car_crashes = sns.load_dataset("car_crashes")
corr = car_crashes.corr()
print(corr)
sns.heatmap(corr,annot=True,linewidths=.5,cmap="YlGnBu")
```

## Facet Grid

```python
# Facer grid forms a matric of panels defined by row and column by dividing the variables.
# Due to panels, a single plot looks like multiple plots. It is very helpfuul to analyze all combinations
# in two discrete variables.
import seaborn as sns
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time")
g.map(sns.scatterplot, "total_bill", "tip")
```

```python
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time", row="sex")
g.map(sns.scatterplot, "total_bill", "tip")
```

## Pair Plot

```python
# Created a grid of Axes such that each numeric variables in data will be shared across the y-axis across
# a single row and the x-axex across a single column
# The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal
# distributino of the data in each column
# It produces a matrix of relationships between each in your data for an instant examination of our data
import seaborn as sns
from matplotlib import pyplot as plt
df = sns.load_dataset('iris')
sns.set_style("ticks")
print(df.head())
sns.pairplot(df, hue='species', diag_kind="kde",
             kind="scatter", palette="husl")
plt.show()
```
